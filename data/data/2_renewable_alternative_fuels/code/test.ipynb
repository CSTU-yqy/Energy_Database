```python
def get_pvtable(year,month,table):
    if not os.path.exists('../data/%s/%s/pv_table%s.xlsx'%(year,month,table)):
        url = 'https://www.eia.gov/renewable/monthly/solar_photo/archive/%s%s/xls/pv_table%s.xlsx'%(month,year,table)
        response = requests.get(url)

        if not os.path.exists('../data/%s'%year):
            os.mkdir('../data/%s'%year)
            
        if not os.path.exists('../data/%s/%s'%(year,month)):
            os.mkdir('../data/%s/%s'%(year,month))
            
        if response.status_code == 200:
            with open('../data/%s/%s/pv_table%s.xlsx'%(year,month,table), 'wb') as f:
                f.write(response.content)
            print("%s %s %sdownloaded"%(year,month,table))
            time.sleep(10)
        else:
            print("request failed :", response.status_code)
```

```python
for i in range(20):
    for year in range(2017,2024):
        for month in ['january','february','march','april','may','june','july','august','september','october','november','december']:
            for table in range(1,10):
                try:
                    get_pvtable(year,month,table)
                except:
                    pass
            
```
